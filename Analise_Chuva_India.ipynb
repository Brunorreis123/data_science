```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
dados_india = pd.read_csv('India_rainfall.csv')

dados_india.head()
```

```python
# Calcular a soma das chuvas por mês
rainfall_sum = dados_india[['Actual Rainfall: JUN', 
                            'Actual Rainfall: JUL', 
                            'Actual Rainfall: AUG', 
                            'Actual Rainfall: SEPT']].sum()

# Criar um gráfico de pizza
plt.pie(rainfall_sum, labels=rainfall_sum.index, autopct='%1.1f%%')
plt.title('Distribution of Rainfall by Month')
plt.axis('equal')  # Para manter o gráfico de pizza redondo
plt.show()
```

```python
# Calcular a soma das chuvas por mês
rainfall_sum = dados_india[['Actual Rainfall: JUN', 
                            'Actual Rainfall: JUL', 
                            'Actual Rainfall: AUG', 
                            'Actual Rainfall: SEPT']].sum()

# Criar um gráfico de barras
bar_width = 0.4  # Largura das barras
index = np.arange(len(rainfall_sum))

plt.bar(index, rainfall_sum, width=bar_width)
plt.xlabel('Months')
plt.ylabel('Rainfall')
plt.title('Distribution of Rainfall by Month')
plt.xticks(index, rainfall_sum.index)
plt.show()
```

```python
# Selecionar o mês desejado para análise (por exemplo, 'Actual Rainfall: JUN')
mes = 'Actual Rainfall: JUN'
dados_mes = dados_india[mes]

# Criar o histograma
plt.hist(dados_mes, bins=10, edgecolor='black')

# Configurar os rótulos e título do gráfico
plt.xlabel('Chuva')
plt.ylabel('Frequência')
plt.title('Distribuição da Chuva em ' + mes)

# Exibir o histograma
plt.show()
```

```python
# Selecionar as colunas de chuva mensal
colunas_chuva = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']

# Calcular a soma acumulativa das chuvas por mês
chuva_acumulada = dados_india[colunas_chuva].cumsum(axis=1)

# Criar o gráfico de área
plt.stackplot(dados_india['YEAR'], chuva_acumulada.T, labels=colunas_chuva)

# Configurar os rótulos e título do gráfico
plt.xlabel('Ano')
plt.ylabel('Chuva Acumulada')
plt.title('Distribuição Acumulativa das Chuvas')

# Adicionar a legenda
plt.legend(loc='upper left')

# Exibir o gráfico de área
plt.show()
```

```python
# Selecionar as colunas relevantes para calcular a correlação
colunas_chuva = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']

# Criar um DataFrame contendo apenas as colunas selecionadas
dados_chuva = dados_india[colunas_chuva]

# Calcular a matriz de correlação
matriz_correlacao = dados_chuva.corr()

# Plotar o gráfico de correlação
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', square=True)

# Configurar o título do gráfico
plt.title('Matriz de Correlação - Chuvas na Índia')

# Exibir o gráfico
plt.show()
```

```python
# Selecionar as colunas relevantes para o boxplot
colunas_chuva = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']

# Criar um DataFrame contendo apenas as colunas selecionadas
dados_chuva = dados_india[colunas_chuva]

# Plotar o gráfico de boxplot
plt.boxplot(dados_chuva.values, labels=colunas_chuva)

# Configurar os rótulos e título do gráfico
plt.xlabel('Mês')
plt.ylabel('Chuva')
plt.title('Distribuição da Chuva por Mês')

# Exibir o gráfico
plt.show()
```

```python
# Selecionar as colunas relevantes para o gráfico de linhas
colunas_chuva = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']

# Plotar o gráfico de linhas
for coluna in colunas_chuva:
    plt.plot(dados_india['YEAR'], dados_india[coluna], label=coluna)

# Configurar os rótulos e título do gráfico
plt.xlabel('Ano')
plt.ylabel('Chuva')
plt.title('Chuvas na Índia ao longo dos Anos')

# Adicionar a legenda
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
# Selecionar as colunas relevantes para o gráfico de barras agrupadas
colunas_chuva = ['Actual Rainfall: JUN', 'Actual Rainfall: JUL', 'Actual Rainfall: AUG', 'Actual Rainfall: SEPT']

# Calcular as médias de chuva para cada mês
medias_chuva = dados_india[colunas_chuva].mean()

# Configurar os rótulos dos meses e as posições das barras
meses = ['JUN', 'JUL', 'AUG', 'SEPT']
posicoes = np.arange(len(meses))

# Plotar o gráfico de barras agrupadas
plt.bar(posicoes, medias_chuva, align='center')

# Configurar os rótulos e título do gráfico
plt.xlabel('Mês')
plt.ylabel('Média de Chuva')
plt.title('Média de Chuva por Mês na Índia')

# Configurar as posições e rótulos do eixo x
plt.xticks(posicoes, meses)

# Exibir o gráfico
plt.show()
```
