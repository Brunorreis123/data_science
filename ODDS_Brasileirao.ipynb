```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar apenas as partidas do Corinthians
corinthians_df = df[(df['HomeTeam'] == 'Corinthians') | (df['AwayTeam'] == 'Corinthians')]

# Filtrar as colunas relevantes para a análise
corinthians_odds = corinthians_df[['Date', 'B365H', 'B365A']]

# Converter a coluna 'Date' para o formato de data
corinthians_odds['Date'] = pd.to_datetime(corinthians_odds['Date'])

# Filtrar as partidas entre 2012 e 2022
corinthians_odds = corinthians_odds[(corinthians_odds['Date'].dt.year >= 2012) & (corinthians_odds['Date'].dt.year <= 2022)]

# Ordenar as odds pela data
corinthians_odds = corinthians_odds.sort_values(by='Date')

# Obter os valores máximos e mínimos de odds
max_odds = max(corinthians_odds['B365H'].max(), corinthians_odds['B365A'].max())
min_odds = min(corinthians_odds['B365H'].min(), corinthians_odds['B365A'].min())

# Criar o gráfico de linha
plt.plot(corinthians_odds['Date'], corinthians_odds['B365H'], marker='o', label='Odds Casa')
plt.plot(corinthians_odds['Date'], corinthians_odds['B365A'], marker='o', label='Odds Fora')

plt.xlabel('Ano')
plt.ylabel('Odds')
plt.title(f'Evolução das Odds do Corinthians (2012-2022)\nMaior Odds: {max_odds:.2f} - Menor Odds: {min_odds:.2f}')
plt.legend()

plt.xticks(rotation=45)
plt.show()
```
